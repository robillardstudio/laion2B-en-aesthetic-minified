Read the first parquet file

```python
import pandas as pd

# Replace 'your_file.parquet' with the path to your snappy.parquet file
df = pd.read_parquet('part-00000-9230b837-b1e0-4254-8b88-ed2976e9cee9-c000.snappy.parquet', engine='pyarrow') 
# or use engine='fastparquet' if you prefer

df.head()
```

```python
rows, columns = df.shape
print(f"Number of rows: {rows}, Number of columns: {columns}")
```

Create new dataframe with a sample of 10 000 rows

```python
# Assuming 'df' is your original DataFrame
sampled_df = df.sample(n=10000, random_state=1)  # Sample 10,000 rows

# Optional: If you want to reset the index of the sampled DataFrame
sampled_df.reset_index(drop=True, inplace=True)
```

Download images from the URL column

```python
import requests
import os
from urllib.parse import urlparse
import pandas as pd

def is_valid_url(url):
    """Check if the URL is valid."""
    parsed = urlparse(url)
    return bool(parsed.netloc) and bool(parsed.scheme)


# Assuming df is your DataFrame and 'URL' is the column with the image URLs
for index, row in sampled_df.iterrows():
    url = row['URL']
    
    file_path = os.path.join('images', f'image_{index}.jpg')

    if not is_valid_url(url):
        print(f"Invalid URL at index {index}: {url}")
        continue

    # Check if the image is already downloaded
    if os.path.exists(file_path):
        print(f"Image already exists at index {index}")
        continue

    try:
        response = requests.get(url, timeout=10)  # 10 seconds timeout

        if response.status_code == 200:

            with open(file_path, 'wb') as file:
                file.write(response.content)
            print(f"Downloaded image at index {index}")
        else:
            print(f"Failed to download image at index {index}: Status code {response.status_code}")
    except requests.RequestException as e:
        print(f"Error downloading image at index {index}: {e}")
```
